```python
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
import numpy as np
import pandas as pd
from scipy.io import arff
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("data.csv")
for column in df.columns:
  df[column], uniques = pd.factorize(df[column])
df = df.to_numpy()
train, test = train_test_split(df, train_size=0.75)
X = train[:, :-1]
y = train[:, -1]
X_test = test[:, :-1]
y_test = test[:, -1]
clf = KNeighborsClassifier(n_neighbors=5, weights='uniform', metric='minkowski')
clf.fit(X, y)
print("train accuracy: ", clf.score(X, y))
print("test accuracy: ", clf.score(X_test, y_test))
```

```python
df = pd.read_csv("data.csv")
for column in df.columns:
  df[column], _ = pd.factorize(df[column])
df = df.to_numpy()
train, test = train_test_split(df, train_size=0.75)
X = train[:, :-1]
y = train[:, -1]
X_test = test[:, :-1]
y_test = test[:, -1]
weights = ['uniform', 'distance']
for i in range(1, 11):
  for weight in weights:
    for p in range(1, 3):
      clf = KNeighborsClassifier(n_neighbors=i, weights=weight, p=p)
      clf.fit(X, y)
      print("weight: ", weight, ", k: ", i, ", p: ", p)
      print("train accuracy: ", clf.score(X, y))
      print("test accuracy: ", clf.score(X_test, y_test))
```
