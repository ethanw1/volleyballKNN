```python
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
import numpy as np
import pandas as pd
from scipy.io import arff
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("data.csv")
for column in df.columns:
  df[column], uniques = pd.factorize(df[column])
train, test = train_test_split(df, train_size=0.75)
X = train.drop('SET_REGION', axis=1)
X = X.to_numpy()
y = train['SET_REGION']
y = y.to_numpy()
X_test = test.drop('SET_REGION', axis=1)
X_test = X_test.to_numpy()
y_test = test['SET_REGION']
y_test = y_test.to_numpy()
clf = KNeighborsClassifier(n_neighbors=5, weights='uniform', metric='minkowski')
clf.fit(X, y)
print("train accuracy: ", clf.score(X, y))
print("test accuracy: ", clf.score(X_test, y_test))
```

```python
df = pd.read_csv("data.csv")
for column in df.columns:
  df[column], _ = pd.factorize(df[column])
train, test = train_test_split(df, train_size=0.75)
X = train.drop('SET_REGION', axis=1)
X = X.to_numpy()
y = train['SET_REGION']
y = y.to_numpy()
X_test = test.drop('SET_REGION', axis=1)
X_test = X_test.to_numpy()
y_test = test['SET_REGION']
y_test = y_test.to_numpy()
weights = ['uniform', 'distance']
for i in range(1, 11):
  for weight in weights:
    for p in range(1, 3):
      clf = KNeighborsClassifier(n_neighbors=i, weights=weight, p=p)
      clf.fit(X, y)
      print("weight: ", weight, ", k: ", i, ", p: ", p)
      print("train accuracy: ", clf.score(X, y))
      print("test accuracy: ", clf.score(X_test, y_test))
```
