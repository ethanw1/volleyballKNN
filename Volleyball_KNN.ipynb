```python
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
import numpy as np
import pandas as pd
from scipy.io import arff
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("data.csv")
for column in df.columns:
  df[column], uniques = pd.factorize(df[column])
train, test = train_test_split(df, train_size=0.75)
X = train.drop('SET_REGION', axis=1)
X = X.to_numpy()
y = train['SET_REGION']
y = y.to_numpy()
X_test = test.drop('SET_REGION', axis=1)
X_test = X_test.to_numpy()
y_test = test['SET_REGION']
y_test = y_test.to_numpy()
clf = KNeighborsClassifier(n_neighbors=5, weights='uniform', metric='minkowski')
clf.fit(X, y)
print("train accuracy: ", clf.score(X, y))
print("test accuracy: ", clf.score(X_test, y_test))
```

```python
def prep_dataset(df, columns_to_drop):
  df = df.drop(columns_to_drop, axis=1)
  for column in df.columns:
    df[column], _ = pd.factorize(df[column])
  return df

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
def prep_dataset2(df, columns_to_drop):
  data = df.drop(columns_to_drop, axis=1)
  # for column in df.columns:
  #   encoder = LabelEncoder()
  #   encoder.fit(df[column])
  #   df[column] = encoder.transform(df[column])
  enc = OneHotEncoder(handle_unknown='ignore')
  categorical_cols = ["SET_REGION", "BYU_FL_Position", "BYU_FM_Position", "BYU_FR_Position", "BYU_BL_Position",
              "BYU_BM_Position", "BYU_BR_Position", "OPP_FL_Position", "OPP_FM_Position", "OPP_FR_Position",
              "OPP_BL_Position", "OPP_BM_Position", "OPP_BR_Position"]
  enc.fit(data[categorical_cols])
  #save the encoder
  import pickle
  pickle.dump(enc, open("one_hot_encoder.pkl", "wb"))
  #%%
  # apply the encoder into a dataframe
  one_hot = enc.transform(data[categorical_cols]).toarray()
  onehot_df = pd.DataFrame(one_hot, columns=enc.get_feature_names_out(categorical_cols))
  numerical = data.drop(columns=categorical_cols, axis=1)
  onehot_df.index = numerical.index
  data = pd.concat([numerical, onehot_df], axis=1)
  data.head()
  return data

def prep_dataset_my_columns(df):
  columns_to_drop = [
    'OPP_BR_number','OPP_BR_ID','OPP_BR_Height','OPP_BR_BS','OPP_BR_BA','OPP_BR_BT',
    'OPP_BR_BE','OPP_BR_SA','OPP_BR_SE','OPP_BM_number','OPP_BM_ID','OPP_BM_Height',
    'OPP_BM_BS','OPP_BM_BA','OPP_BM_BT','OPP_BM_BE','OPP_BM_SA','OPP_BM_SE','OPP_BL_number','OPP_BL_ID',
    'OPP_BL_Height','OPP_BL_BS','OPP_BL_BA','OPP_BL_BT','OPP_BL_BE','OPP_BL_SA','OPP_BL_SE'
  ]
  return prep_dataset2(df, columns_to_drop)
```
